```python
# ALUMNA: Ana Cristina Torres Faustino. ALUMNA: Eunice Jadiel Corro Rivera
class ArbolTrie:
    class Nodo:
        def __init__(self):
            self.palabra = False  #indica si el nodo representa una palabra
            self.caracter = None  #almacena el caracter asociado al nodo
            self.hijos = {}

    def __init__(self):
        self.raiz = self.Nodo()
        self.codigo_caracter = {}  #diccionario para el codigo por cada letra

    def append2(self, letra, codigo): #insertando las palabras y bit
        nodoact = self.raiz  #nodo actual
        for bit in codigo:
            if bit not in nodoact.hijos:  #si el bit no esta en el diccionario
                nodoact.hijos[bit] = self.Nodo()  #agrega el bit al diccionario
            nodoact = nodoact.hijos[bit]
        nodoact.palabra = True  #termina la palabra
        nodoact.caracter = letra
        self.codigo_caracter[letra] = codigo  #almacenar el codigo por cada letra

    def buscar(self, codigo):
        nodoact = self.raiz  #nodo actual
        for bit in codigo:
            if bit not in nodoact.hijos:
                return None
            nodoact = nodoact.hijos[bit]  #avanza al siguiente nodo
        return nodoact if nodoact.palabra else None  #retorna el nodo si la palabra es encontrada

    def codificar(self, texto):
        tcodif = ""  #texto codificado
        for caracter in texto:
            if caracter in self.codigo_caracter:
                tcodif += self.codigo_caracter[caracter]  #agrega el codigo por cada letra
        return tcodif

    def decodificar(self, ctexto):
        nodoact = self.raiz  #nodo actual
        tdescifrado = ""  #texto decifrado
        for bit in ctexto:
            if bit in nodoact.hijos:
                nodoact = nodoact.hijos[bit]  #avanza al siguiente nodo
                if nodoact.palabra:
                    tdescifrado += nodoact.caracter  #agrega la letra al texto decifrado
                    nodoact = self.raiz  #reiniciar al nodo raiz
        return tdescifrado

if __name__ == "__main__":
    arbol = ArbolTrie()
    arbol.append2("A", "000")
    arbol.append2("B", "001")
    arbol.append2("C", "10")
    arbol.append2("D", "11")
    arbol.append2("R", "01")

    ctexto = arbol.codificar("ARBRCAARBDRA")
    dtexto = arbol.decodificar(ctexto)
    print("Texto Codificado:", ctexto)
    print("Texto Descifrado:", dtexto)
arboltrie = ArbolTrie()
def cargar():

    valores = {
        ' ': '01', ',': '10000110', '.': '1010001', 'a': '1001', 'b': '00001010',
        'c': '1000001', 'd': '101010', 'e': '11', 'f': '10000100', 'g': '00001001',
        'h': '101001', 'i': '00010', 'j': '10000001', 'k': '10000000', 'l': '101011',
        'm': '1010000', 'n': '00000', '単': '10000101', 'o': '1011', 'p': '00001011',
        'q': '00001110', 'r': '10001', 's': '00011', 't': '001', 'u': '0000110',
        'v': '00001111', 'y': '00001000', 'z': '10000111'
    }
    for letra, codigo in valores.items():
        arboltrie.append2(letra, codigo)
    return arboltrie

def decodificar2(arboltrie):


    cb = "000001011010000101100001101010100001010100001011000110100011111000110100010110101110010100100010111000110001100101000001011010000101100001101010101011010011001000000011011101000101000001011010001110111010000101100011011000001000011010010000000110110100011110100001011100011011000010110000110000111011011110101101000111011101011011100000010000110000000110010000010100111101011101101100011110100001011001101110100010100001100000001000010110001011010001111011001100000111100011000001100101100101101011101101100000110101110011000110111010001011100000011010111011011011000111000001000011010001101101000100000000011000100001100111011110101101101100110001101110100010100001011101110001000011100000110110111101011011001101000010111000101000001011011100011010000101111100010000101111001000011010110111000000100000100110101000010111000011001000000001001110000001101000000010010011001101000000001011101110000011011101000101111010110110111010100001010110110010000100100001101001100011010101001010001100001100100010000000001100110010000000111011010101100000001100011011011010101110101101100000110011000100001010101100000011010000100100011011010011011000001010101011101000101100011011100000011011011100011011110101101101110101000010101101000010000100000000011000110001000101010101011101000101111010110110011010000101110001100001100100001011100110101100010101010101101000010000100011101110101110111010001011000001100100000000111001101010101101101010110110111010100001010011000110000110010011101100110100001011101000101100000110010000000011100110101010110110101011011001101000010011000110000110010011101101110101000010101110100010110101110101100001101100001111110100100010111010000000010111011100001100110101110101100001101100001111110100100010111010000000010111011101000101000010000100001100000001101010000101001010011000100010000110011101110000000110001110100110111010101011000111000011001001100010001000011001110100001011101110001010011011101010100101101011100101001000101110001100011001100001100100110001000100001100111011010101100011101010110110100000001001101001100100011001100101111010110110101110110000101010111000011001101010101110001101000000010101000010110001101000010000110101011000110000101111100010011001101000010110001101100000110110000001000011000000010010001000001001100011101110000000110001110110101110110001101101110000001101100011101000101000010110001011101010100011001000111000011001101001101110100000000101010001110001101100000110111010000101101000010110001011101010100011001000111000011001101010000011010001101100011010000100001000010111010111100000101100011011010101101110000010000011011000001011100001100110000011010011011100000110010000001110000001111010110110010001010001111000011001101010101100000101010110110000011010011011100000110010000001101011100100011010000101100010111010101000110010001101101010110100001011100011011000000011011101000101000111011101011111010101001101010110001101000011100000110110110100110110000100001100011110000011010011001100001111001000000100001000011001000010001110001011000000100001100000000110010000101010010000001000110000110000110110001101100011001100011011000111010001010001110111010111110101010011010101100011010000111000001101101110000001111010110100001010110001110110100001001000011010011000110101010010000001111010110110001000011000001001000101010101011010001110111000110101010111010001010001110111010111110101010011010101100011010000101110011000110010100011000101110100000000101110001111010001010001110111010111110101010011010101100011010001100010000000110010000101110110000100010111010001011000001000011011100010000101110110001101100000110111010000101101000011000000011010000100110001010000111110111000110011000011110000110011100000001100011110000011010011011100000110011010101011100001100110000100000011010001000101011000111011101000101000111011101011000100011001100010001010011010000110000000111011001001100110101010110001101000010111011100010111101011011001101000010111000110000110010000101110111000101111010110110111010100001010111010001010000101110111000101101011100100011010000111111000001001000110100011000011010001000010011100000011010011011101000000001010100011100011100001100110000011000100001101000011110010000001101011100100011011000001000100000110101010100110101011000111000011001001101110001000011111011000111010001011100000011110101101100000110111000110011000011110110000001100110001100011001000100000100110010100011101110101100010001100110001000101001101000011000000011101001101110101010111010001011010010000110111010111010111001000110100011000100000001100000110111010000000010111001100001011001010000111000001101110101010010000001100000110111010000101101110000001111010110110010000100100001101001100001100111000000111101011011000010010110000010101010111010001010001110111010111011010000110000001001010000111110111000011110000110011001011010111011011010111110000001101100011100001100100011000101110100000000101110001110110010110101110110110101111100000011011000110110101110010110110001000001001101110000110011001100000110111010000000010111001100001011001010000100001101001100110000011101000101000101000100000100100001101001101011010000111000001101101111010110110000010000110111010111010111011011001011010111011000110110100110111010000000010101000110110001110100010100011101110101110110100001100000010010100001111101110000111011010000110110011000110001110000101010010011001011100011001110110011000110100001001100001111011000000111000110000101100010000001011000111011011010101101000011111110101110101110110110001110011000110111000001111010100001010101010110100001000011110001000101000011110011010101011010000111000001101101101000011010000101110110000000001001101110100010110101110110001101000111110000011011000110100001111000101100000001101100011010000010110100001011000011011101010110000001000111110000011001100010110101110110001101101000010011000111000110110000001000011000001001101100011101100011101000101000010000111101011011000001101110001100110000111101100000010000101111100011010000100100000111000001110110000100100011100011100000110110111000000100011000011001100000110011000110000011110101101101010110110010000100110110001100110110100001011101110001000011100000110110111000111001010000111110111000011101110001101111010110110011000110100001001011010000100100011010010001011100010000010010110101011011010111011000110110011000110001101100001000101100011101000101101000000010000010010000110111010111000011001101000011011001100000100001101110001101010101101101010110100100010011010101100011000010110000110110001101101010111010110100011101110101101000010000110101011101011010000101110111010110000111110111000011001100100000001110001101101010110110101110010110100000001000011101000010010110101100001100000010011000011001001000011010100000000101010010110101011010000110000000100011000011011100001011011011001101000010111000110110001110111010001011001101000010111000110000110011001101011111000000110010110100000001001000111110001011010101101000110000110000110100001011100010001010100001110001101100011011100011100000110111010000000010101000110110001110000110010000100001111010100001010000100000101000001100100000101010101110100001110000110011010100001010000010011001010000110000001001010000111111100011010101001101010011000001101110100001011010000110000000100011101100001011101011101110100010110101011000110000101100001101100011011010101110101101100110100001011100011000011001101011100101001000101110001100011001101000101101010110001100001011000011011000110110101011011010111001010010001011100011000110011000011001001101110101010111010001"
    text = arboltrie.decodificar(cb)
    print("Texto codificado")
    print(text)

arboltrie = cargar()
decodificar2(arboltrie)
```
